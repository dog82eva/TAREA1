```python
# Cambiar la ciudad de todos los artistas a "Nueva York"
for artista in artistas_y_ciudades:
    artistas_y_ciudades[artista] = "Nueva York"

# Mostrar la nueva relación de artistas y ciudades
print("Estos son los conciertos para esta semana: " + str(artistas_y_ciudades))
```
